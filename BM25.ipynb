# 中文斷詞

```python
from udicOpenData.stopwords import *

# 分詞加空白
def split_passage(passage):
    word_and_tag = list(rmsw(passage, flag=True))
    _temp = list(filter(lambda x: x[1]!='nr' and x[1]!='ng', word_and_tag))
    # print(_temp)
    text = " ".join(list(map(lambda x:x[0], _temp)))
    return text
```

```python
for item in train_data:
    item['contents'] = split_passage(item['fact'])
    del item['fact']

for item in test_data:
    item['contents'] = split_passage(item['fact'])
    del item['fact']
```

# 使用反向索引

```python
import json

with open('dataset/test_data.json', 'r', encoding='utf-8') as f:
    test_data = json.load(f)
```

```python
test_data[0]['fact']
```

```python
from pyserini.search.lucene import LuceneSearcher
import time
import json

# 改成反向索引的資料夾
searcher = LuceneSearcher('indexes')

# 設定成中文
searcher.set_language('zh')

search_key = test_data[0]['fact']

# 長度只能吃到1024
if len(search_key) > 1024:
    search_key = search_key[:1024]

# 計算下一次查詢要多久
start = time.time()
hits = searcher.search(search_key, k=5) # 預設: k=10
end = time.time()

sentence_list = [] #五個取平均，算MSE

print("排名\t文章編號\t\t\tBM25分數\t刑期(天)\t\t內容")
for i in range(len(hits)):
    idx = hits[i].docid
    raw_doc = searcher.doc(idx).raw()
    content = json.loads(raw_doc)['contents']
    sentence = json.loads(raw_doc)['sentance']
    sentence_list.append(sentence)
    print(f'{i+1}\t{idx}\t{hits[i].score:.5f}\t{sentence}\t\t{content}')

print("\n執行時間：%f 秒" % (end - start))
print(f'Pred: {int(sum(sentence_list)/len(sentence_list))}')
```

```python
from pyserini.search.lucene import LuceneSearcher
from sklearn.metrics import mean_squared_error, mean_absolute_error
import time
import json

# 改成反向索引的資料夾
searcher = LuceneSearcher('indexes')

# 設定成中文
searcher.set_language('zh')

with open('dataset/test_data.json', 'r', encoding='utf-8') as f:
    dataset = json.load(f)

y_ans = []
y_pred = []

for data in dataset:
    search_key = data['fact']
    # print(search_key)

    # 長度只能吃到1024
    if len(search_key) > 1024:
        search_key = search_key[:1024]

    # 計算下一次查詢要多久
    start = time.time()
    hits = searcher.search(search_key, k=5) # 預設:k=10
    end = time.time()

    sentence_list = [] #五個取平均，算MSE

    # print("排名\t文章編號\t\t\tBM25分數\t刑期(天)\t\t內容")
    for i in range(len(hits)):
        idx = hits[i].docid
        raw_doc = searcher.doc(idx).raw()
        content = json.loads(raw_doc)['contents']
        sentence = json.loads(raw_doc)['sentance']
        sentence_list.append(sentence)
        # print(f'{i+1}\t{idx}\t{hits[i].score:.5f}\t{sentence}\t\t{content}')

    ans = data['sentance']
    pred = int(sum(sentence_list)/len(sentence_list))
    y_ans.append(ans)
    y_pred.append(pred)
    # print(f'Ans: {ans}, Pred: {pred}\n' + '='*200)

print('\n'+'='*200)
print(f'Ans: {y_ans}')
print(f'Pred: {y_pred}')

mse_linear = mean_squared_error(y_ans, y_pred)
rmse_linear = mean_squared_error(y_ans, y_pred, squared=False)
mae_linear = mean_absolute_error(y_ans, y_pred)

print('\n'+'='*200)
print("MSE:", mse_linear)
print("RMSE:", rmse_linear)
print("MAE:", mae_linear)
```
